```python
from athena.src.services.download.data import Data
from athena.src.services.pipeline.pipeline import Pipeline

rules = {
    "domains": ["stocks", "fixed etfs"]
}

download = Data(rules = rules, period = "4y")
data = download.get_data()

filters = {

    "volume": {
        "num_tickers": 100,
        "months": 1
    },

    "momentum": {
        "num_tickers": 8,
        "months": 6
    }

}

optimizer = {
    "hrp": {
        "covariance_function": "sample_cov",
        "linkage_method": "single"
    }
}

pipeline = Pipeline(data = data, filters = filters, optimizer = optimizer)
weights = pipeline.get_weights()
```
