```python
returns = new.pct_change()

performance = pd.DataFrame(columns = returns.columns, index = ["Mean", "Std", "Sharpe"])

for column in returns.columns:
    
    performance[column]["Mean"]   = np.mean(returns[column])
    performance[column]["Std"]    = np.std(returns[column])
    performance[column]["Sharpe"] = (np.mean(returns[column]) - 0.000162)/np.std(returns[column])       # 0.000162 a.d. = 0.06 a.a.


performance
```

```python
returns = new.pct_change()
```

```python
hrp = HRPOpt(returns = returns.drop("TECK11", axis = 1))

hrp.optimize()
hrp.clean_weights()
hrp.portfolio_performance()
weights = hrp.clean_weights()
weights
```

```python
maximum = max([data[ticker].values[-1]/weights[ticker] for ticker in list(weights.keys())])

shares  = dict()

for ticker in list(weights.keys()):

    shares[ticker] = np.round(maximum*weights[ticker]/data[ticker].values[-1])

total = sum([shares[ticker]*data[ticker].values[-1] for ticker in list(weights.keys())])

print(f"--------- Total {total} ---------")
print(shares)
```
