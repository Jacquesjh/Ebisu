```python
from data import Data
import json
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
with open("stock_sectors.json") as f:
    brasil_sectors = json.load(f)

rules = {
    "domains": ["stocks"]
}

download = Data(rules = rules, period = "4y")
data = download.get_data()["Close"]

ibov = yf.download("^BVSP", period = "4y")["Close"]
```

```python
df = pd.DataFrame(index = ["mean", "std"])

for sector in brasil_sectors:

    sector_mean = data[brasil_sectors[sector]].pct_change().mean().mean()
    sector_std  = data[brasil_sectors[sector]].pct_change().std().mean()
    sector_dist = np.random.normal(sector_mean, sector_std, 200)

    df[sector] = [sector_mean, sector_std]

    ibov_mean = ibov.pct_change().mean()
    ibov_std  = ibov.pct_change().std()
    ibov_dist = np.random.normal(ibov_mean, ibov_std, 200)
```

```python
std = df.loc["std"].sort_values()
std
```

```python
import plotly.figure_factory as ff

dists = [np.random.normal(df[sector]["mean"], df[sector]["std"], 200) for sector in df.columns]
fig = ff.create_distplot(dists, group_labels = df.columns)
fig.show()
```
